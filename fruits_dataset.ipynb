```
fruits = ['Mango', 'Apple', 'Banana', 'Pomegranate', 'Watermelon', 'Strawberry', 'Chikoo', 'Grapes', 'Guava', 'Pineapple']
```

```
for fruit in fruits:
  folder = fruit
  url_file = fruit.lower()+'.csv'
  path = Path('fruits_data')
  dest = path/folder
  dest.mkdir(parents=True, exist_ok=True)
  download_images(url_file, dest, max_pics=2000)
```
